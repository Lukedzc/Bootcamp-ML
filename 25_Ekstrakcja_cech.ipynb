Załadowanie danych

```python

def fetch_financial_data(company='AMZN'): #tiker do pobierania danych - notowania spólki Amazon -
    """
    This function fetches stock market quotations. #źródło danych stock - tiker
    """
    import pandas_datareader.data as web
    return web.DataReader(name=company, data_source='stooq')

df_raw = fetch_financial_data()
df_raw.head()
```

```python

df = df_raw.copy()
df = df[:5]
df.info()
```

Dyskretyzacja zmiennej ciągłej

```python
df = pd.DataFrame(data={'height': [175., 178.5, 185., 191., 184.5, 183., 168.]})
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=3)
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195)) # tworzymy trzy przedziały
df
```

```python
df['height_cat'] = pd.cut(x=df.height, bins=(160, 175, 180, 195), labels=['small', 'medium', 'high']) # parametr labels przekazuje ['small', 'medium', 'high']
df
```

```python

pd.get_dummies(df, drop_first=True, prefix='height') # używamy parametr: get_dummies
```

Ekstrakcja cech

```python

df = pd.DataFrame(data={'lang': [['PL', 'ENG'], ['GER', 'ENG', 'PL', 'FRA'], ['RUS']]})
df
```

```python

df['lang_number'] = df['lang'].apply(len) # biblioteka pandas posiada metodę applay, która pozwoli przekazać funkcję którą możemy użyć na konkretnym wierszu (w tym przypadku jest to funkcja len - zwraca dł. listy)
df
```

```python
df['PL_flag'] = df['lang'].apply(lambda x: 1 if 'PL' in x else 0) # zmienna: df['PL_flag'] - metoda apply; funkaja lambda - zwróci 1 jeżeli pl jest w x w przciwnym razie zwróci -0
df
```
