Import bibliotek

```python

import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

Wygenerowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}#przekazanie danych w postaci s≈Çownika

df_raw = pd.DataFrame(data=data)  #tworzenie obiektu dataframe
df_raw
```

Utworzenie kopii danych


```python

df = df_raw.copy()
df.info() #podstawowe informacje na temat obiektu
```

Zmiana typu danych i wstƒôpna eksploracja

```python

for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

Standaryzacja - StandardScaler

std() - pandas nieobciƒÖ≈ºony

std() - numpy obciƒÖ≈ºony

Kolumny z warto≈õciami numerycznymi powinny zachowywaƒá balans (jedne warto≈õci nie powinny przewy≈ºszaƒá znaczƒÖco drugich) ‚Äì nale≈ºy przeprowadziƒá standaryzacjƒô.
(obliczamy ≈õredniƒÖ z danych odejmujemy od danych i dzielimy przez odchylenie standardowe)
Odchylenie standardowe jest estymatorem obciƒÖ≈ºonym wariancji
Zaimplementowana postaƒá estymatora mo≈ºe siƒô r√≥≈ºniƒá z uwagi na to z jakiej biblioteki zostanie zaimportowany: pandas lub numpy

```python
# ≈õrednia oraz odchylenie z biblioteki pandas
print(f"{df['price']}\n")
print(f"≈örednia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std() #od ka≈ºdej warto≈õci odejmujemy ≈õredniƒÖ i dzielimy przez odchylenie standardowe; dane zosta≈Çy ustandaryzowane i skupiajƒÖ siƒô wok√≥≈Ç zera
```

```python
#funkcjƒô mo≈ºna napisaƒá samodzielnie:
def standardize(x):
    return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
#lub zaimportowaƒá  z biblioteki sklearn; warto≈õci bƒôdƒÖ siƒô r√≥≈ºniƒá wynika to z kwesti estymatora obciƒÖ≈ºonego oraz nieobciƒÖ≈ºonego wariancji; dlatego warto≈õƒá odcylenia standardowego ulega zmianie
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
# polecana klasa: StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # tworzymy instancjƒô klasy
scaler.fit(df[['price']]) # u≈ºywamy metody fit kt√≥ra pozwoli dopasowaƒá siƒô do kolumny price (obliczy statystyki, kt√≥re musimy wyko≈ºystaƒá do standaryzacji)
scaler.transform(df[['price']]) # metoda transform - przekszta≈Çci dane
```

Kolejny etap to przekszta≈Çcenie danych numerycznych (kolumny numeryczne) - przeprowadziƒá standaryzacjƒô

```python
# tworzymy instancjƒô klasy StandardScaler i przekazujemy dwie kolumny: price i weight
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```

Przygotowanie danych do modelu

```python
df = df_raw.copy()
df
```

```python

le = LabelEncoder() # tworzymy klasƒô LabelEncoder kt√≥ra zakoduje zmiennƒÖ docelowƒÖ 'bought'

df['bought'] = le.fit_transform(df['bought'])

scaler = StandardScaler() # skalowanie zmiennych numerycznych
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(data=df, drop_first=True) # kodowanie zmiennych kategorycznych
df
```
