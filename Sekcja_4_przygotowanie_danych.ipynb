PREPROCESSING DANYCH


Import bibliotek

# Nowa sekcja

```python

import numpy as np
import pandas as pd
import sklearn

sklearn.__version__
```

Wygenerowanie danych

```python
data = {
    'size': ['XL', 'L', 'M', 'L', 'M'],
    'color': ['red', 'green', 'blue', 'green', 'red'],
    'gender': ['female', 'male', 'male', 'female', 'female'],
    'price': [199.0, 89.0, 99.0, 129.0, 79.0],
    'weight': [500, 450, 300, 380, 410],
    'bought': ['yes', 'no', 'yes', 'no', 'yes']
}#przekazanie danych w postaci słownika

df_raw = pd.DataFrame(data=data)  #tworzenie obiektu dataframe
df_raw
```

Utworzenie kopii danych


```python

df = df_raw.copy()
df.info() #podstawowe informacje na temat obiektu
```

Zmiana typu danych i wstępna eksploracja

```python

for col in ['size', 'color', 'gender', 'bought']:
    df[col] = df[col].astype('category')

df['weight'] = df['weight'].astype('float')

df.info()
```

PRZYGOTOWANIE MODELU DO UCZENIA MASZYNOWEGO

```python

from sklearn.preprocessing import LabelEncoder
#mapowanie zmiennych 'bought'
le = LabelEncoder() #instancja klasy - przypisujemy do le =; klasa labelencoder jest tworzona głównie do przekształcania zmiennych docelowych, nie używamy dla zmiennych opisujących
le.fit(df['bought']) # fit dopasuje encoder do zmiennej bought
le.transform(df['bought']) #mapowanie, transformowanie danych
```

```python

le.fit_transform(df['bought']) #sybsza forma powyższego zapisu
```

```python

le.classes_ #wyświetla jak mapowanie przebiegało
```

```python
df['bought'] = le.fit_transform(df['bought']) #przypisujemy obiekt datafreme do modelu
df
```

```python
le.inverse_transform(df['bought']) #powrót do pierwotnej postaci
```

```python

df['bought'] = le.inverse_transform(df['bought'])
df
```

Standaryzacja - StandardScaler

std() - pandas nieobciążony

std() - numpy obciążony

Kolumny z wartościami numerycznymi powinny zachowywać balans (jedne wartości nie powinny przewyższać znacząco drugich) – należy przeprowadzić standaryzację.
(obliczamy średnią z danych odejmujemy od danych i dzielimy przez odchylenie standardowe)
Odchylenie standardowe jest estymatorem obciążonym wariancji
Zaimplementowana postać estymatora może się różnić z uwagi na to z jakiej biblioteki zostanie zaimportowany: pandas lub numpy

```python
# średnia oraz odchylenie z biblioteki pandas
print(f"{df['price']}\n")
print(f"Średnia: {df['price'].mean()}")
print(f"Odchylenie standardowe: {df['price'].std():.4f}")
```

```python
(df['price'] - df['price'].mean()) / df['price'].std() #od każdej wartości odejmujemy średnią i dzielimy przez odchylenie standardowe; dane zostały ustandaryzowane i skupiają się wokół zera
```

```python
#funkcję można napisać samodzielnie:
def standardize(x):
    return (x - x.mean()) / x.std()

standardize(df['price'])
```

```python
#lub zaimportować  z biblioteki sklearn; wartości będą się różnić wynika to z kwesti estymatora obciążonego oraz nieobciążonego wariancji; dlatego wartość odcylenia standardowego ulega zmianie
from sklearn.preprocessing import scale

scale(df['price'])
```

```python
# polecana klasa: StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler() # tworzymy instancję klasy
scaler.fit(df[['price']]) # używamy metody fit która pozwoli dopasować się do kolumny price (obliczy statystyki, które musimy wykożystać do standaryzacji)
scaler.transform(df[['price']]) # metoda transform - przekształci dane
```

Kolejny etap to przekształcenie danych numerycznych (kolumny numeryczne) - przeprowadzić standaryzację

```python
# tworzymy instancję klasy StandardScaler i przekazujemy dwie kolumny: price i weight
scaler = StandardScaler()
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])
df
```

Przygotowanie danych do modelu

```python
df = df_raw.copy()
df
```

```python

le = LabelEncoder() # tworzymy klasę LabelEncoder która zakoduje zmienną docelową 'bought'

df['bought'] = le.fit_transform(df['bought'])

scaler = StandardScaler() # skalowanie zmiennych numerycznych
df[['price', 'weight']] = scaler.fit_transform(df[['price', 'weight']])

df = pd.get_dummies(data=df, drop_first=True) # kodowanie zmiennych kategorycznych
df
```
